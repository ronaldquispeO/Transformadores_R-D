```python
# SOLUCIÓN PARA EL ERROR DE DUPLICADOS
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 1. Primero, obtener la última fecha de cada serie (sin duplicados)
df_ultima_fecha = df_R.sort_values('FECHA').groupby('SERIE', as_index=False).last()

print("Últimos registros por serie:")
print(df_ultima_fecha[['SERIE', 'FECHA']])

# 2. Identificar columnas H POS
columnas_h_pos = [col for col in df_ultima_fecha.columns if 'H POS' in col and 'R 75°C' in col]
print(f"\nColumnas H POS encontradas: {len(columnas_h_pos)}")

# 3. Elegir UNA serie específica (la primera disponible)
serie_elegida = df_ultima_fecha['SERIE'].iloc[0]
print(f"\nSerie elegida: {serie_elegida}")

# 4. Obtener datos de esa serie (sin duplicados)
datos_serie = df_ultima_fecha[df_ultima_fecha['SERIE'] == serie_elegida].iloc[0]

# 5. Crear el gráfico de barras
plt.figure(figsize=(14, 6))

valores = []
posiciones_nombres = []
posiciones_numeros = []

# Recoger los datos válidos
for col in columnas_h_pos:
    valor = datos_serie[col]
    if not pd.isna(valor):
        valores.append(valor)
        # Extraer el número de posición
        num_pos = col.split('POS')[1].split(' ')[0]
        posiciones_nombres.append(f'POS {num_pos}')
        posiciones_numeros.append(int(num_pos))

# Ordenar por número de posición
datos_ordenados = sorted(zip(posiciones_numeros, posiciones_nombres, valores))
pos_nums_ordenados, pos_nombres_ordenados, valores_ordenados = zip(*datos_ordenados)

# Crear el gráfico de barras
barras = plt.bar(pos_nombres_ordenados, valores_ordenados, 
                 color='steelblue', alpha=0.7, edgecolor='navy', linewidth=1)

# Personalizar el gráfico
plt.xlabel('Posiciones')
plt.ylabel('Resistencia [mΩ]')
plt.title(f'Resistencias - Serie {serie_elegida}\n(Fecha: {datos_serie["FECHA"]})')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3, axis='y')

# Añadir valores encima de las barras
for barra, valor in zip(barras, valores_ordenados):
    altura = barra.get_height()
    plt.text(barra.get_x() + barra.get_width()/2, altura + 1, 
             f'{valor:.1f}', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

# 6. Mostrar los datos utilizados
print(f"\nDatos utilizados para {serie_elegida}:")
print(f"Fecha: {datos_serie['FECHA']}")
print("\nValores por posición:")
for pos, val in zip(pos_nombres_ordenados, valores_ordenados):
    print(f"  {pos}: {val:.2f} mΩ")
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 1. VERIFICAR QUÉ COLUMNAS TIENES EN df (TODAS LAS FASES)
print("🔍 COLUMNAS DISPONIBLES EN df (TODAS LAS FASES):")

# Buscar columnas para R, S, T
columnas_r = [col for col in df.columns if 'R 75°C' in col and 'H POS' in col]
columnas_s = [col for col in df.columns if 'S 75°C' in col and 'H POS' in col]
columnas_t = [col for col in df.columns if 'T 75°C' in col and 'H POS' in col]

print(f"Columnas Fase R: {len(columnas_r)}")
print(f"Columnas Fase S: {len(columnas_s)}")
print(f"Columnas Fase T: {len(columnas_t)}")

# Mostrar ejemplos
print("\nEjemplos de columnas R:", columnas_r[:3])
print("Ejemplos de columnas S:", columnas_s[:3])
print("Ejemplos de columnas T:", columnas_t[:3])

# 2. OBTENER ÚLTIMOS DATOS POR SERIE (USANDO df COMPLETO)
df_ultima_fecha = df.sort_values('FECHA').groupby('SERIE', as_index=False).last()

# 3. ELEGIR UN TRANSFORMADOR ESPECÍFICO
serie_elegida = df_ultima_fecha['SERIE'].iloc[0]
datos_serie = df_ultima_fecha[df_ultima_fecha['SERIE'] == serie_elegida].iloc[0]

print(f"\n🎯 ANALIZANDO TRANSFORMADOR: {serie_elegida}")
print(f"📅 Fecha de medición: {datos_serie['FECHA']}")

# 4. ORGANIZAR DATOS PARA LAS 3 FASES
datos_organizados = {'R': {}, 'S': {}, 'T': {}}

def extraer_numero_posicion(nombre_columna):
    """Extraer el número de posición del nombre de columna"""
    try:
        return int(nombre_columna.split('POS')[1].split(' ')[0])
    except:
        return None

# Llenar datos para cada fase
for fase, columnas in zip(['R', 'S', 'T'], [columnas_r, columnas_s, columnas_t]):
    for columna in columnas:
        posicion = extraer_numero_posicion(columna)
        if posicion and not pd.isna(datos_serie[columna]):
            datos_organizados[fase][posicion] = datos_serie[columna]

# 5. CREAR GRÁFICO DE BARRAS AGRUPADAS (R, S, T)
plt.figure(figsize=(16, 8))

# Encontrar todas las posiciones que tienen datos en cualquier fase
todas_posiciones = set()
for fase in ['R', 'S', 'T']:
    todas_posiciones.update(datos_organizados[fase].keys())

todas_posiciones = sorted(todas_posiciones)

if todas_posiciones:
    # Configurar el gráfico de barras agrupadas
    x = np.arange(len(todas_posiciones))
    ancho_barra = 0.25
    
    # Colores para R, S, T
    colores = {'R': '#FF6B6B', 'S': '#4ECDC4', 'T': '#45B7D1'}
    
    # Crear barras para cada fase
    for i, fase in enumerate(['R', 'S', 'T']):
        valores = [datos_organizados[fase].get(pos, 0) for pos in todas_posiciones]
        
        # Solo graficar si hay valores diferentes de cero
        if any(valores):
            barras = plt.bar(x + i * ancho_barra, valores, width=ancho_barra, 
                            label=f'Fase {fase}', color=colores[fase], alpha=0.8,
                            edgecolor='black', linewidth=0.5)
            
            # Añadir valores encima de las barras
            for barra, valor in zip(barras, valores):
                if valor > 0:
                    plt.text(barra.get_x() + barra.get_width()/2, barra.get_height() + 0.5, 
                            f'{valor:.1f}', ha='center', va='bottom', fontsize=8, fontweight='bold')
    
    # Personalizar el gráfico
    plt.xlabel('Posiciones del Transformador')
    plt.ylabel('Resistencia [mΩ]')
    plt.title(f'RESISTENCIAS POR FASE - Transformador {serie_elegida}\n(Fecha: {datos_serie["FECHA"]})')
    plt.xticks(x + ancho_barra, [f'POS {pos}' for pos in todas_posiciones], rotation=45)
    plt.legend()
    plt.grid(True, alpha=0.3, axis='y')
    
    plt.tight_layout()
    plt.show()
    
    # 6. TABLA DE DATOS COMPLETA (R, S, T)
    print(f"\n📊 DATOS DETALLADOS - {serie_elegida}:")
    print("Posición | Fase R [mΩ] | Fase S [mΩ] | Fase T [mΩ]")
    print("-" * 55)
    
    for pos in todas_posiciones:
        r_val = datos_organizados['R'].get(pos, 'N/A')
        s_val = datos_organizados['S'].get(pos, 'N/A') 
        t_val = datos_organizados['T'].get(pos, 'N/A')
        
        # Formatear valores (manejar números y 'N/A')
        r_str = f"{r_val:10.1f}" if r_val != 'N/A' else "       N/A"
        s_str = f"{s_val:10.1f}" if s_val != 'N/A' else "       N/A" 
        t_str = f"{t_val:10.1f}" if t_val != 'N/A' else "       N/A"
        
        print(f"POS {pos:2d}    | {r_str} | {s_str} | {t_str}")

else:
    print("❌ No se encontraron datos de posiciones para este transformador")

# 7. GRÁFICO DE LÍNEAS COMPARATIVO
plt.figure(figsize=(14, 7))

for fase in ['R', 'S', 'T']:
    posiciones = sorted(datos_organizados[fase].keys())
    valores = [datos_organizados[fase][pos] for pos in posiciones]
    
    if valores:
        plt.plot(posiciones, valores, marker='o', linewidth=2, markersize=6,
                label=f'Fase {fase}', color=colores[fase])
        
        # Etiquetas de puntos
        for pos, val in zip(posiciones, valores):
            plt.annotate(f'{val:.1f}', (pos, val), textcoords="offset points", 
                        xytext=(0,10), ha='center', fontsize=8)

plt.xlabel('Posición')
plt.ylabel('Resistencia [mΩ]')
plt.title(f'COMPARACIÓN DE FASES - Transformador {serie_elegida}')
plt.legend()
plt.grid(True, alpha=0.3)
if todas_posiciones:
    plt.xticks(todas_posiciones)
plt.tight_layout()
plt.show()

# 8. RESUMEN ESTADÍSTICO POR FASE
print(f"\n📈 RESUMEN ESTADÍSTICO - {serie_elegida}:")
for fase in ['R', 'S', 'T']:
    valores_fase = list(datos_organizados[fase].values())
    if valores_fase:
        print(f"Fase {fase}: {len(valores_fase)} posiciones | "
              f"Min: {min(valores_fase):.1f} | Max: {max(valores_fase):.1f} | "
              f"Promedio: {np.mean(valores_fase):.1f} | Desv: {np.std(valores_fase):.1f}")
    else:
        print(f"Fase {fase}: Sin datos")

print(f"\n✅ Proceso completado usando df COMPLETO con todas las fases!")
```
