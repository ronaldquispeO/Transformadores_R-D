## FACTOR DE POTENCIA

```python

# Columnas base que deben estar en ambos (SERIE, FECHA)
cols_base = ["SERIE", "FECHA"]

# Columnas con % (factor de potencia)
cols_fp_h = [c for c in df_h.columns if c.endswith("%")]
df_h_fp = df_h[cols_base + cols_fp_h]

cols_fp_x = [c for c in df_x.columns if c.endswith("%")]
df_x_fp = df_x[cols_base + cols_fp_x]

cols_fp_y = [c for c in df_y.columns if c.endswith("%")]
df_y_fp = df_y[cols_base + cols_fp_y]
```

```python
df_h_fp['Max_FP_h'] = df_h_fp[cols_fp_h].max(axis=1)
df_x_fp['Max_FP_x'] = df_x_fp[cols_fp_x].max(axis=1)
df_y_fp['Max_FP_y'] = df_y_fp[cols_fp_y].max(axis=1)
df_h_fp.head()
```

```python
# ==========================
# Configuración de reglas
# ==========================

# Pesos (puedes cambiarlos aquí)
pesos = [5, 3, 1]  

# Límites (puedes cambiarlos aquí)
limites = {
    "FP": [1.0, 0.5]  # [Límite superior, Límite inferior]
}

# ==========================
# Función genérica
# ==========================
def asignar_puntaje(df, columna_valor, columna_puntaje, limites, pesos):
    """
    Aplica reglas de puntuación a un DataFrame.

    Parámetros:
    - df: DataFrame al que se aplican los puntajes.
    - columna_valor: Nombre de la columna con los valores numéricos (ej: 'Max_FP').
    - columna_puntaje: Nombre de la columna de salida (ej: 'FPD_Puntaje').
    - limites: Lista [lim_sup, lim_inf] con los valores de referencia.
    - pesos: Lista [peso_alto, peso_medio, peso_bajo] con los puntajes.
    """
    df[columna_puntaje] = np.select(
        [
            df[columna_valor] > limites[0],                               # Mayor al límite superior
            (df[columna_valor] > limites[1]) & (df[columna_valor] <= limites[0]), # Entre lim_inf y lim_sup
            df[columna_valor] <= limites[1]                               # Menor o igual al límite inferior
        ],
        pesos,  # Puntajes configurables
        default=np.nan
    )
    return df

# ==========================
# Aplicación a cada DataFrame
# ==========================

# PARA FP_H
df_h_fp = asignar_puntaje(df_h_fp, 'Max_FP_h', 'FPD_Puntaje_h', limites["FP"], pesos)

# PARA FP_X
df_x_fp = asignar_puntaje(df_x_fp, 'Max_FP_x', 'FPD_Puntaje_x', limites["FP"], pesos)

# PARA FP_Y
df_y_fp = asignar_puntaje(df_y_fp, 'Max_FP_y', 'FPD_Puntaje_y', limites["FP"], pesos)

# ==========================
# Visualización de ejemplo
# ==========================
df_h_fp.head()
```

## CAPACITANCIA DE DEVANADOS

```python
# Columnas con pF (capacitancia)
cols_cd_h = [c for c in df_h.columns if c.endswith("pF")]
df_h_cd = df_h[cols_base + cols_cd_h]

cols_cd_x = [c for c in df_x.columns if c.endswith("pF")]
df_x_cd = df_x[cols_base + cols_cd_x]

cols_cd_y = [c for c in df_y.columns if c.endswith('pF')]
df_y_cd = df_y[cols_base + cols_cd_y]
```

### Cálculo delta H

```python
# Asegurarnos que FECHA sea tipo datetime
df_h_cd["FECHA"] = pd.to_datetime(df_h_cd["FECHA"])

# Columnas de capacitancia (las que terminan en pF)
cap_cols = [c for c in df_h_cd.columns if c.endswith("pF")]

# Crear nuevas columnas con la variación porcentual respecto al valor inicial (mínima fecha por SERIE)
for col in cap_cols:
    # serie con el valor inicial por SERIE
    ref = df_h_cd.groupby("SERIE").apply(lambda g: g.loc[g["FECHA"].idxmin(), col])
    ref_mapped = df_h_cd["SERIE"].map(ref)

    # cálculo normal
    delta = abs((df_h_cd[col] - ref_mapped) / ref_mapped) * 100

    # reemplazar inf por NaN (división por cero)
    delta = delta.replace([np.inf, -np.inf], np.nan)

    # si alguno es NaN o el valor/ref_mapped es 0 -> NaN
    mask = df_h_cd[col].isna() | ref_mapped.isna() | (ref_mapped == 0) | (df_h_cd[col] == 0)
    delta[mask] = np.nan

    df_h_cd[f"{col}_Delta"] = delta
```
