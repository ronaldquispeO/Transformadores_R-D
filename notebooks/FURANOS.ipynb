```python
import pandas as pd
import numpy as np
address1 = 'C:/Users/RONALD Q/OneDrive - LUZ DEL SUR S.A.A/Documentos/Estudios de Ingreso/ProyectoRyD/Basededatos/FURANOS.xlsx'
address2 = 'C:/Users/roquispec/OneDrive - LUZ DEL SUR S.A.A/Documentos/Estudios de Ingreso/ProyectoRyD/Basededatos/FURANOS.xlsx'
df = pd.read_excel(address2,header=1)
```

```python
df.head(10)
```

```python
df = df.dropna(axis=1,how='all')
df['Fecha'] = pd.to_datetime(df['Fecha'])
df.head(10)
```

```python
df['FUR'] = np.select(
    [
        df['2-Furfuraldehido (FAL, ppb)'] > 5000,                               # Mayor 5000
        (df['2-Furfuraldehido (FAL, ppb)'] > 1000) & (df['2-Furfuraldehido (FAL, ppb)'] <= 5000), # Entre 5000% y 1000%
        (df['2-Furfuraldehido (FAL, ppb)'] > 500) & (df['2-Furfuraldehido (FAL, ppb)'] <= 1000), # Entre 1000% y 500%
        (df['2-Furfuraldehido (FAL, ppb)'] > 100) & (df['2-Furfuraldehido (FAL, ppb)'] <= 500), # Entre 500% y 100%
        df['2-Furfuraldehido (FAL, ppb)'] <= 100
                                        # Menor o igual a 100%
    ],
    [5, 4, 3, 2, 1],  # Puntajes
    default=np.nan
)
df = df.rename(columns={'Fecha': 'FECHA DE MUESTRA'})
df = df[['SERIE','FECHA DE MUESTRA','FUR']]
```
