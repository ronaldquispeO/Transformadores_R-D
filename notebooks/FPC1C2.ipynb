### FACTOR DE POTENCIA DE BUSHING C1

```python
# ==========================
# Configuración de reglas
# ==========================

# Pesos (puedes cambiarlos aquí)
pesos = [5, 3, 1]  

# Límites (puedes cambiarlos aquí)
limites = {
    "FP": [1.0, 0.5]  # [Límite superior, Límite inferior]
}

# ==========================
# Función genérica
# ==========================
def asignar_puntaje(df, columna_valor, columna_puntaje, limites, pesos):
    """
    Aplica reglas de puntuación a un DataFrame.

    Parámetros:
    - df: DataFrame al que se aplican los puntajes.
    - columna_valor: Nombre de la columna con los valores numéricos (ej: 'Max_FP').
    - columna_puntaje: Nombre de la columna de salida (ej: 'FPD_Puntaje').
    - limites: Lista [lim_sup, lim_inf] con los valores de referencia.
    - pesos: Lista [peso_alto, peso_medio, peso_bajo] con los puntajes.
    """
    df[columna_puntaje] = np.select(
        [
            df[columna_valor] > limites[0],                               # Mayor al límite superior
            (df[columna_valor] > limites[1]) & (df[columna_valor] <= limites[0]), # Entre lim_inf y lim_sup
            df[columna_valor] <= limites[1]                               # Menor o igual al límite inferior
        ],
        pesos,  # Puntajes configurables
        default=np.nan
    )
    return df

# ==========================
# Aplicación a cada DataFrame
# ==========================

# PARA FP_H
df_fp_c1 = asignar_puntaje(df_fp_c1, 'Max_FP_C1', 'FPBC1', limites["FP"], pesos)

# ==========================
# Visualización de ejemplo
# ==========================
df_fp_c1
```

# Solo Capacitancias

```python
cols_c = [col for col in df.columns if col.endswith('pF')]
df_c = df[cols_base+cols_c]
df_c
```

```python
cols_c_c1 = [col for col in df_c.columns if col.startswith('C1')]
df_c_c1 = df_c[cols_base+cols_c_c1]
df_c_c1
```

```python

# Asegurarnos que FECHA sea tipo datetime
df_c_c1["FECHA"] = pd.to_datetime(df_c_c1["FECHA"])

# Columnas de capacitancia (las que terminan en pF)
cap_cols = [c for c in df_c_c1.columns if c.endswith("pF")]

# Crear nuevas columnas con la variación porcentual respecto al valor inicial (mínima fecha por SERIE)
for col in cap_cols:
    # serie con el valor inicial por SERIE
    ref = df_c_c1.groupby("SERIE").apply(lambda g: g.loc[g["FECHA"].idxmin(), col])
    ref_mapped = df_c_c1["SERIE"].map(ref)

    # cálculo normal
    delta = abs((df_c_c1[col] - ref_mapped) / ref_mapped) * 100

    # reemplazar inf por NaN (división por cero)
    delta = delta.replace([np.inf, -np.inf], np.nan)

    # si alguno es NaN o el valor/ref_mapped es 0 -> NaN
    mask = df_c_c1[col].isna() | ref_mapped.isna() | (ref_mapped == 0) | (df_c_c1[col] == 0)
    delta[mask] = np.nan

    df_c_c1[f"{col}_Delta"] = delta
df_c_c1.head()
```

```python

cols_c_c1_delta = [col for col in df_c_c1.columns if col.endswith('Delta')]

df_c_c1['Max_C_C1'] = df_c_c1[cols_c_c1_delta].max(axis=1)
df_c_c1
```
