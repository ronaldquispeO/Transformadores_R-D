```python
df_fp['Max_FP'] = df_fp[[c for c in df_fp.columns if c.endswith('%')]].max(axis=1)
df_fp.head()
```

```python
import numpy as np

# ==========================
# Configuración de reglas
# ==========================

# Pesos 
pesos = [5, 3, 1]  

# Límites 
limites = {
    "FP": [1.0, 0.5]  # [Límite superior, Límite inferior]
}

# ==========================
# Función genérica
# ==========================
def asignar_puntaje(df, columna_valor, columna_puntaje, limites, pesos):
    """
    Aplica reglas de puntuación a un DataFrame.

    Parámetros:
    - df: DataFrame al que se aplican los puntajes.
    - columna_valor: Nombre de la columna con los valores numéricos (ej: 'Max_FP').
    - columna_puntaje: Nombre de la columna de salida (ej: 'FP').
    - limites: Lista [lim_sup, lim_inf] con los valores de referencia.
    - pesos: Lista [peso_alto, peso_medio, peso_bajo] con los puntajes.
    """
    df[columna_puntaje] = np.select(
        [
            df[columna_valor] > limites[0],                               # Mayor al límite superior
            (df[columna_valor] > limites[1]) & (df[columna_valor] <= limites[0]), # Entre lim_inf y lim_sup
            df[columna_valor] <= limites[1]                               # Menor o igual al límite inferior
        ],
        pesos,  # Puntajes configurables
        default=np.nan
    )
    return df

# ==========================
# Aplicación a df_fp
# ==========================
df_fp = asignar_puntaje(df_fp, 'Max_FP', 'FP', limites["FP"], pesos)

# ==========================
# Visualización de ejemplo
# ==========================
df_fp

```

# CAPACITANCIA DE DEVANADOS

```python
# Seleccionar columnas que terminan con 'pF'
cols_cd = ["SERIE", "FECHA"] + [c for c in df.columns if c.endswith("pF")]
df_cd = df[cols_cd]
df_cd
```

```python
# Asegurarnos que FECHA sea tipo datetime
df_cd["FECHA"] = pd.to_datetime(df_cd["FECHA"], errors="coerce")

# Columnas de capacitancia (las que terminan en pF)
cap_cols = [c for c in df_cd.columns if c.endswith("pF")]

# Crear nuevas columnas con la variación porcentual respecto al valor inicial (mínima fecha por SERIE)
for col in cap_cols:
    # valor inicial por SERIE
    ref = df_cd.groupby("SERIE").apply(lambda g: g.loc[g["FECHA"].idxmin(), col])
    ref_mapped = df_cd["SERIE"].map(ref)

    # cálculo normal
    delta = abs((df_cd[col] - ref_mapped) / ref_mapped) * 100

    # reemplazar inf por NaN
    delta = delta.replace([np.inf, -np.inf], np.nan)

    # condiciones para dejar como NaN
    mask = df_cd[col].isna() | ref_mapped.isna() | (ref_mapped == 0) | (df_cd[col] == 0)
    delta[mask] = np.nan

    # guardar columna de delta
    df_cd[f"{col}_Delta"] = delta
df_cd
```

```python
cap_cols_delta_y = [c for c in df_cd.columns if c.endswith("Delta")]
df_cd['Max_CD'] = df_cd[cap_cols_delta_y].max(axis=1)
df_cd.head()
```

```python
# PARA CD H
df_cd['CD'] = np.select(
    [
        df_cd['Max_CD'] > 10,                               # Mayor a 10%
        (df_cd['Max_CD'] > 5) & (df_cd['Max_CD'] <= 10), # Entre 5% y 10%
        df_cd['Max_CD'] <= 5                             # Menor o igual a 5%
    ],
    [5, 3, 1],  # Puntajes
    default=np.nan
)

df_cd.head()
```

```python
df_cd = df_cd.drop(columns='Max_CD')
```
